# Data augmentation

```python
datagen = ImageDataGenerator(
        rotation_range=35,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.4,
        horizontal_flip=True,
        fill_mode='nearest')
```

# total blackrot

```python
os.chdir(r"C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/dataset_image_augmented/Blackrot")
for path in os.listdir():
    img = load_img(f"{path}")
    x = img_to_array(img)    # this is a Numpy array with shape (3, 150, 150)
    x = x.reshape((1,) + x.shape)
    i = 0
    for batch in datagen.flow(x, batch_size=1,
            save_to_dir=".", 
            save_prefix= 'blackrot', 
            save_format='jpg'):
        i += 1
        if i == 2:     ## creates 10 image form 1 image 
            break 
```

# NLM filtrado datos aumentados

```python
os.chdir(r"C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing")
images_path = "./dataset_image_augmented"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path + "/*"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the images path
        image = cv2.imread(file)
        # denoising images
        dst = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
        b,g,r = cv2.split(dst)
        rgb_dst = cv2.merge([r,g,b])
        # save images
        save_path = './Dataset_vid_prueba_den/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.jpg'),
                    cv2.cvtColor(rgb_dst,cv2.COLOR_RGB2BGR))
        

        # show loop update
        i += 1
```
