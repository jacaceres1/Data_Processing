```python
import cv2
import os
import numpy as np
import glob
from sklearn.preprocessing import MinMaxScaler
import h5py
import mahotas as mt
```

# Extraer las caracteristicas

```python
images_path = "C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Dataset_bg_normales"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
def extract_features(image):
	# calculate haralick texture features for 4 types of adjacency
	textures = mt.features.haralick(image)

	# take the mean of it and return it
	ht_mean  = textures.mean(axis=0)
	return ht_mean
```

```python
# declarar los arreglos para almacenar la informaci√≥n
textural_features = []
textural_features_labels   = []
```

```python
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path + "/*.png"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the training image
        image = cv2.imread(file)

        # convert the image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # extract haralick texture from the image
        features = extract_features(gray)

        # append the feature vector and label
        textural_features.append(features)
        textural_features_labels.append(cur_label)

        # show loop update
        i += 1
```

# Generar el dataset

```python
def replace(my_list, my_dict):
    return [x if x not in my_dict else my_dict[x] for x in my_list]

labels_dict = {'Blackrot': 0, 'Esca': 1, 'Leaf_blight': 2,'Healthy':3}
target = replace(textural_features_labels, labels_dict)
```

```python
scaler            = MinMaxScaler(feature_range=(0, 1))
rescaled_features = scaler.fit_transform(textural_features)
```

```python
h5f_data = h5py.File('C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Datos_transfromados_Sin/textural_features.h5', 'w')
h5f_data.create_dataset('Pest_Textural_Features', data=np.array(rescaled_features))

h5f_label = h5py.File('C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Datos_transfromados_Sin/textural_features_labels.h5', 'w')
h5f_label.create_dataset('Pest_Textural_Features_Labels', data=np.array(target))

h5f_data.close()
h5f_label.close()
```
