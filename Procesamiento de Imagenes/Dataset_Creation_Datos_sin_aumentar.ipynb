```python
import cv2
import os
import numpy as np
import glob
from sklearn.preprocessing import MinMaxScaler
import h5py
import mahotas as mt
```

# Extraer caracteristicas de colores

```python
images_path = "C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Dataset_vid_prueba_densidad"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
#Crear el histograma para la obtención de los colores
def fd_histogram(image, mask=None):
    # convert the image to HSV color-space
    image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    # compute the color histogram
    hist  = cv2.calcHist([image], [0, 1, 2], None, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    # normalize the histogram
    cv2.normalize(hist, hist)
    # return the histogram
    return hist.flatten()
```

```python
#Creación de los arreglos para almacenar las caracteristicas
color_features = []
color_features_labels   = []
```

```python
print("[STATUS] Started extracting color features..")
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path + "/*.jpg"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the training image
        image = cv2.imread(file)

        # extract color feature from the image
        bins = 8
        hist = fd_histogram(image)

        # append the feature vector and label
        color_features.append(hist)
        color_features_labels.append(cur_label)

        # show loop update
        i += 1
```

# Creación del data set

```python
# Obtención del tamaño de los datos de las caracteristicas
print("[STATUS] color feature vector size {}".format(np.array(color_features).shape))

#obtener el tamaño de las variables de entrenamiento
print("[STATUS] training Labels {}".format(np.array(color_features_labels).shape))
```

```python
def replace(my_list, my_dict):
    return [x if x not in my_dict else my_dict[x] for x in my_list]

labels_dict = {'Blackrot': 0, 'Esca': 1, 'Leaf_blight': 2,'Healthy':3}
target = replace(color_features_labels, labels_dict)

print("[STATUS] training labels encoded...")
```

```python
# scale features in the range (0-1)
scaler            = MinMaxScaler(feature_range=(0, 1))
rescaled_features = scaler.fit_transform(color_features)
print("[STATUS] feature vector normalized...")
print("[STATUS] target labels: {}".format(np.array(target)))
print("[STATUS] target labels shape: {}".format(np.array(target).shape))
```

```python
h5f_data = h5py.File('C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/color_features.h5', 'w')
h5f_data.create_dataset('Leaf_grapewine_color_features', data=np.array(rescaled_features))

h5f_label = h5py.File('C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/color_features_labels.h5', 'w')
h5f_label.create_dataset('Leaf_grapewine_color_features_Labels', data=np.array(target))

h5f_data.close()
h5f_label.close()

print("[STATUS] end of training..")
```
