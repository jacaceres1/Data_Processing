```python
import os
import glob
import cv2
import rembg
from rembg import remove
```

```python
"""
    images_path = se deterina el directorio donde se almacenan las imagenes del dataset
    class_names = se recupera las clases que se encuentran dentro del dataset de imagenes
    Funciones Implementadas:
    listdir() permite listar los elementos que se encuentren dentro de una ruta o directorio especifico
    sort() permite ordenar una lista en orden ascendente o descendente
"""
images_path = "C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Dataset_vid_prueba_bg_lab"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
"""
    for declaracion de clases:
        - Recupera el nombre de los archivos correspondientes con la clase.
        - Se aplica el cambio al nombre del atributo correspondiente
        for: recorre los elementos
            - Realiza la lecutra del directorio correspondiente.
            - Realiza el procedimiento de eliminación del fondo.
            - Realizar el proceso de guardado con nuevo formato y nombre
"""
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path + "/*"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the images path
        image = cv2.imread(file)
        
        #procedimiento
        output = remove(image)

        # save images
        save_path = 'C:/Users/JOSE/Documents/Septimo/Mineria/Image_processing/Dataset_bg_normales/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.png'), output)
        

        # show loop update
        i += 1
```
